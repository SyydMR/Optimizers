```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**2

def func_grad(x):
    return 2 * x
```

```python
def gradient_descent(lr, func_grad):
    x = -10
    x_values = [x]
    for i in range(15):
        x -= lr * func_grad(x)
        x_values.append(x)
    return x_values
```

```python
def search_path(x_values, func):
    x_line = np.arange(-15, 15, 0.01)
    plt.plot(x_line, [func(x) for x in x_line])
    plt.plot(x_values, [func(x) for x in x_values], 'ro-')
```

```python
x_values = gradient_descent(0.1, func_grad)
search_path(x_values, func)
```
